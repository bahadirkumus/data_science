## İleri Birleştirme İşlemleri

```python
df1 = pandas.DataFrame({'calisanlar': ['Ali', 'Veli', 'Ayse', 'Fatma'],
                    'grup': ['Muhasebe', 'Muhendislik', 'Muhendislik', 'İK']})
```

```python
df1
```

```python
df2 = pandas.DataFrame({'calisanlar': ['Ayse', 'Ali', 'Veli', 'Fatma'],
                    'ilk_giris': [2010, 2009, 2014, 2019]})
```

```python
df2
```

```python
pandas.merge(df1, df2)
```

```python
df3 = pandas.merge(df1, df2)
```

```python
df3
```

```python
df4 = pandas.DataFrame({'grup': ['Muhasebe', 'Muhendislik', 'İK'],
                    'mudur': ['Caner', 'Mustafa', 'Berkcan']})
```

```python
df4
```

```python
pandas.merge(df3,df4)
```

```python
df5 = pandas.DataFrame({'grup': ['Muhasebe', 'Muhasebe',
                              'Muhendislik', 'Muhendislik', 'İK', 'İK'],
                    'yetenekler': ['matematik', 'excel', 'kodlama', 'linux',
                               'excel', 'yonetim']})
```

```python
df5
```

```python
pandas.merge(df1, df5)
```
