## İleri Toplulaştırma İşlemleri (Aggregate, filter, transform, apply)

```python
df3 = pd.DataFrame({'gruplar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'degisken1': [10, 43, 65, 56, 34, 23],
                   'degisken2': [100, 542, 345, 676, 563, 342]},
                   columns = ['gruplar', 'degisken1', 'degisken2'])
```

```python
df3
```

```python
#aggregate
```

```python
df3.groupby('gruplar').mean()
```

```python
df3.groupby('gruplar').aggregate(['min', np.median, max])
```

```python
df3.groupby('gruplar').aggregate({'degisken1': min, 'degisken2': max})
```

```python
#filter
```

```python
def filter_func(x):
    return x['degisken1'].std() > 9
```

```python
df3.groupby('gruplar').filter(filter_func)
```

```python
#transform
```

```python
df3['degisken1']*9
```

```python
df3.transform(lambda x: x-x.mean())
```

```python
df3.transform(lambda x: x-x.mean() / x.std())
```

```python
#apply
```

```python
df4 = pd.DataFrame({'degisken1': [10, 43, 65, 56, 34, 23],
                   'degisken2': [100, 542, 345, 676, 563, 342]},
                   columns = ['degisken1', 'degisken2'])
```

```python
df4.apply(np.sum)
```

```python
df4.apply(np.mean)
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.groupby('sex')['survived'].mean()
```

```python
titanic.groupby(['sex', 'class'])[['survived']].mean()
```

```python
titanic.groupby(['sex', 'class'])[['survived']].aggregate('mean').unstack()
```

```python
#pivot
```

```python
titanic.pivot_table('survived', index = 'sex', columns = 'class')
```

```python
titanic.age.head()
```

```python
age = pd.cut(titanic['age'], [0, 18, 90])
```

```python
age
```
